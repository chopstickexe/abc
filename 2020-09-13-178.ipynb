 # B


```python
inputs = [int(x) for x in input().split()]
a = inputs[0]
b = inputs[1]
c = inputs[2]
d = inputs[3]

# Avoid using 0
if a == 0 and a + 1 <= b:
    a += 1
if b == 0 and b - 1 >= a:
    b -= 1
if c == 0 and c + 1 <= d:
    c += 1
if d == 0 and d - 1 >= c:
    d -= 1

if (a > 0 and b > 0 and c < 0 and d < 0) or (a < 0 and b < 0 and c > 0 and d > 0):
    # Return the max negative value
    print(-(min(abs(a), abs(b)) * min(abs(c), abs(d))))
else:
    # Return the max positive value
    print(max(a * c, b * d))
```

 # C

 愚直に解いてみたが、まあTLE。（計算も間違ってる）

```python
import math
import sys

import numpy as np
from scipy.special import comb

N = int(input())
if N < 2:
    print(0)
    sys.exit()

# Num of arrays consist of 0 and 9 only
# (but subtract a number of 0-only and 9-only arrays)
ans = np.power(2, N) - 2
x = 1
while 2 + x <= N:
    # Fill x sheets by 1~8
    ans += (np.power(2, (N - x)) - 2) * comb(N, x, exact=True) * np.power(8, x)
    x += 1
print(ans % (10 ** 9 + 7))
```

 ## 解説を踏まえたコード
 Python選択の場合で396ms, PyPy3の場合で286msで終わる。


```python
def powmod(x, y, mod):
    res = pow(x, y) % mod
    return res


N = int(input())
mod = pow(10, 9) + 7
# (x + y) mod z = (x mod z + y mod z) mod z
ans = powmod(10, N, mod) - 2 * powmod(9, N, mod) + powmod(8, N, mod)
ans = ans % mod
print(ans)
```

 ちなみに、Pythonなのでいちいちmod取らずに最後だけmod取るのでも良い


```python
N = int(input())
mod = pow(10, 9) + 7
# (x + y) mod z = (x mod z + y mod z) mod z
ans = pow(10, N) - 2 * pow(9, N) + pow(8, N)
ans = ans % mod
print(ans)
```

 # D

 本番では間に合わなかったが、DPを使って解く問題らしい。

```python
S = int(input())
dp = [0] * (S + 1)
mod = pow(10, 9) + 7

for i in range(0, S + 1):
    if i == 0:
        dp[i] = 1
        continue
    if i < 3:
        # dp[i] = 0
        continue
    dp[i] += (dp[i - 3] + dp[i - 1]) % mod 
print(dp[S])
```
